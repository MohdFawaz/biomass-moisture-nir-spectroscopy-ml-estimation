```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.stats import zscore

# Assuming 'data' is already loaded and 'Unnamed: 1' dropped

# 1. Outlier Detection on Moisture Using IQR Method
Q1 = data['Moisture'].quantile(0.25)
Q3 = data['Moisture'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers in Moisture
moisture_outliers = data[(data['Moisture'] < lower_bound) | (data['Moisture'] > upper_bound)]
print("Moisture Outliers:")
print(moisture_outliers[['Sample ID', 'Moisture']])

# Visualize the Moisture distribution with outliers
plt.figure(figsize=(4,3))
plt.boxplot(data['Moisture'], vert=False)
plt.title('Boxplot of Moisture Content')
plt.xlabel('Moisture')
plt.show()

# 2. Outlier Detection on Spectral Data Using PCA
# Define spectral columns (assuming first column is 'Sample ID' and last is 'Moisture')
spectral_cols = data.columns[1:-1]
X = data[spectral_cols].values

# Apply PCA to reduce dimensions (we use 2 components for visualization and outlier detection)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Compute Euclidean distances from the mean in the PCA space
mean_pca = np.mean(X_pca, axis=0)
distances = np.linalg.norm(X_pca - mean_pca, axis=1)

# Compute z-scores of the distances
z_scores = zscore(distances)

# Flag samples with z-score > 3 as outliers (you can adjust the threshold as needed)
spectral_outliers = data[z_scores > 3]
print("\nSpectral Outliers based on PCA distances:")
print(spectral_outliers[['Sample ID', 'Moisture']])

# Optional: Visualize the PCA results and highlight outliers
plt.figure(figsize=(4,3))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', label='Normal Samples', alpha=0.5)
plt.scatter(X_pca[z_scores > 3, 0], X_pca[z_scores > 3, 1], c='red', label='Outliers', marker='x', s=100)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Spectral Data')
plt.legend()
plt.show()
```

# 2- Data Pre-processing

**Here sample normalization is performed using min-max normalization for each spectrum**

**Then Savitzky-Golay smoothing is applied**

**Thirdly, comparisons of processing techniques on samples were visalized**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# Assuming 'data' is already loaded and the 'Unnamed: 1' column has been dropped
# Define spectral columns (assuming first column is 'Sample ID' and last is 'Moisture')
spectral_cols = data.columns[1:-1]

# 1. Sample Normalization: Using min-max normalization for each spectrum
def normalize_spectrum(row):
    min_val = row.min()
    max_val = row.max()
    return (row - min_val) / (max_val - min_val)

data_normalized = data.copy()
data_normalized[spectral_cols] = data_normalized[spectral_cols].apply(normalize_spectrum, axis=1)

# 2. Savitzky-Golay Smoothing: Choose window length and polynomial order (window must be odd)
window_length = 31  # You can adjust this parameter
polyorder = 3      # You can adjust this parameter

data_smoothed = data_normalized.copy()
data_smoothed[spectral_cols] = data_smoothed[spectral_cols].apply(
    lambda row: pd.Series(savgol_filter(row, window_length, polyorder), index=row.index),
    axis=1
)

# Visualize the effect on a sample spectrum (e.g., sample index 100)
sample_index = 141
wavelengths = [float(col) for col in spectral_cols]

plt.figure(figsize=(6, 3))
plt.plot(wavelengths, data.loc[sample_index, spectral_cols], label="Raw Spectrum", alpha=0.5)
plt.plot(wavelengths, data_normalized.loc[sample_index, spectral_cols], label="Normalized Spectrum", alpha=0.7)
plt.plot(wavelengths, data_smoothed.loc[sample_index, spectral_cols], label="Smoothed Spectrum", linewidth=2)
plt.xlabel("Wavelength")
plt.ylabel("Absorbance")
plt.title("Comparison of Preprocessing Techniques on a Sample Spectrum")
plt.legend()
plt.show()

fig, axs = plt.subplots(3, 1, figsize=(5, 5), sharex=True)

# Raw Spectrum
axs[0].plot(wavelengths, data.loc[sample_index, spectral_cols], color='blue')
axs[0].set_title("Raw Spectrum")

# Normalized Spectrum
axs[1].plot(wavelengths, data_normalized.loc[sample_index, spectral_cols], color='green')
axs[1].set_title("Normalized Spectrum (scaled to [0, 1])")

# Smoothed Spectrum
axs[2].plot(wavelengths, data_smoothed.loc[sample_index, spectral_cols], color='red')
axs[2].set_title("Smoothed Spectrum")

plt.xlabel("Wavelength")
plt.tight_layout()
plt.show()
```

Here we are applying the 3 ML models on the smoothed data (I used 5-fold cross-validation data splitting and shuffling):


*  1- PLS Regression
*  2- Support Vector Regression
*  3- Artificial Neural Networkusing MLP regressor






```python
import numpy as np
import pandas as pd
from sklearn.cross_decomposition import PLSRegression
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

# Prepare the data
# We'll use the preprocessed (smoothed) spectral data as features (X) and Moisture as target (y)
spectral_cols = data_smoothed.columns[1:-1]  # assuming first col is Sample ID and last is Moisture
X = data_smoothed[spectral_cols].values
y = data_smoothed['Moisture'].values

# Setup 5-fold cross-validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# 1. PLS Regression
# Choose a number of components (e.g., 10) - you can adjust this based on your data.
pls = PLSRegression(n_components=10)
y_pred_pls = cross_val_predict(pls, X, y, cv=cv)
r2_pls = r2_score(y, y_pred_pls)
rmse_pls = np.sqrt(mean_squared_error(y, y_pred_pls))

print("PLS Regression Results:")
print("R²: {:.3f}".format(r2_pls))
print("RMSE: {:.3f}".format(rmse_pls))

# 2. Support Vector Regression (SVR)
# We'll build a pipeline that scales the features first, then applies SVR.
# The parameters (e.g., C, gamma) here are examples; you might need to tune them.
svr_pipeline = make_pipeline(StandardScaler(), SVR(kernel='rbf', C=100, gamma=0.01))
y_pred_svr = cross_val_predict(svr_pipeline, X, y, cv=cv)
r2_svr = r2_score(y, y_pred_svr)
rmse_svr = np.sqrt(mean_squared_error(y, y_pred_svr))

print("\nSVR Results:")
print("R²: {:.3f}".format(r2_svr))
print("RMSE: {:.3f}".format(rmse_svr))

# 3. Artificial Neural Network (ANN) using MLPRegressor
# Here, we use a simple one-hidden-layer network. You can experiment with the architecture.
mlp = MLPRegressor(hidden_layer_sizes=(100,), activation='relu', solver='adam',
                   max_iter=1000, random_state=42)
y_pred_ann = cross_val_predict(mlp, X, y, cv=cv)
r2_ann = r2_score(y, y_pred_ann)
rmse_ann = np.sqrt(mean_squared_error(y, y_pred_ann))

print("\nANN (MLPRegressor) Results:")
print("R²: {:.3f}".format(r2_ann))
print("RMSE: {:.3f}".format(rmse_ann))

#Plotting actual vs predicted moisture for one of the models (e.g., PLS)
plt.figure(figsize=(4, 4))
plt.scatter(y, y_pred_pls, alpha=0.7, edgecolor='k')
plt.xlabel("Actual Moisture")
plt.ylabel("Predicted Moisture")
plt.title("PLS Regression: Actual vs. Predicted Moisture")
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linestyle='--')  # Diagonal line
plt.show()

# Plotting actual vs predicted moisture for SVR
plt.figure(figsize=(4, 4))
plt.scatter(y, y_pred_svr, alpha=0.7, edgecolor='k')
plt.xlabel("Actual Moisture")
plt.ylabel("Predicted Moisture")
plt.title("SVR: Actual vs. Predicted Moisture")
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linestyle='--')  # Diagonal line
plt.show()

# Plotting actual vs predicted moisture for ANN
plt.figure(figsize=(4, 4))
plt.scatter(y, y_pred_ann, alpha=0.7, edgecolor='k')
plt.xlabel("Actual Moisture")
plt.ylabel("Predicted Moisture")
plt.title("ANN: Actual vs. Predicted Moisture")
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linestyle='--')  # Diagonal line
plt.show()
```

**Here we run the 3 ML models on the 3 kinds of data: Raw, normalized, and smoothed**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_decomposition import PLSRegression
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Set up 5-fold cross-validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Prepare the feature matrices for each pre-processing method
spectral_cols = data.columns[1:-1]  # assuming first column is Sample ID and last is Moisture

# Raw data features
X_raw = data[spectral_cols].values
# Normalized data features (already computed in data_normalized)
X_norm = data_normalized[spectral_cols].values
# Smoothed data features (already computed in data_smoothed)
X_smooth = data_smoothed[spectral_cols].values

# Target variable (Moisture) remains the same
y = data['Moisture'].values

# Dictionary of preprocessed data arrays
preproc_data = {
    'Raw': X_raw,
    'Normalized': X_norm,
    'Smoothed': X_smooth
}

# Define the models we want to compare
models = {
    'PLS': PLSRegression(n_components=10),  # you can adjust n_components
    'SVR': make_pipeline(StandardScaler(), SVR(kernel='rbf', C=100, gamma=0.01)),
    'ANN': MLPRegressor(hidden_layer_sizes=(100,), activation='relu', solver='adam',
                        max_iter=1000, random_state=42)
}

# We'll store the performance results here:
results = {}

# Loop over each pre-processing method and each model
for method, X in preproc_data.items():
    results[method] = {}
    print(f"--- Preprocessing: {method} ---")
    for model_name, model in models.items():
        # Use cross_val_predict to get predictions on the validation folds
        y_pred = cross_val_predict(model, X, y, cv=cv)
        # Calculate R² and RMSE
        r2 = r2_score(y, y_pred)
        rmse = np.sqrt(mean_squared_error(y, y_pred))
        results[method][model_name] = {'R2': r2, 'RMSE': rmse}
        print(f"{model_name}: R² = {r2:.3f}, RMSE = {rmse:.3f}")
    print()

# Now, let's prepare the data for plotting the results.
methods = list(results.keys())
model_names = list(models.keys())

# Create arrays for R² and RMSE with dimensions: [n_methods x n_models]
r2_values = np.zeros((len(methods), len(model_names)))
rmse_values = np.zeros((len(methods), len(model_names)))

for i, method in enumerate(methods):
    for j, model_name in enumerate(model_names):
        r2_values[i, j] = results[method][model_name]['R2']
        rmse_values[i, j] = results[method][model_name]['RMSE']

# Plot bar charts for R² scores
fig, ax = plt.subplots(figsize=(5, 3))
width = 0.25  # width of each bar
x = np.arange(len(methods))
for j, model_name in enumerate(model_names):
    ax.bar(x + j * width, r2_values[:, j], width, label=model_name)
ax.set_xticks(x + width)
ax.set_xticklabels(methods)
ax.set_ylabel("R² Score")
ax.set_title("R² Scores Across Preprocessing Methods and Models")
ax.legend()
plt.show()

# Plot bar charts for RMSE
fig, ax = plt.subplots(figsize=(5, 3))
for j, model_name in enumerate(model_names):
    ax.bar(x + j * width, rmse_values[:, j], width, label=model_name)
ax.set_xticks(x + width)
ax.set_xticklabels(methods)
ax.set_ylabel("RMSE")
ax.set_title("RMSE Across Preprocessing Methods and Models")
ax.legend()
plt.show()

```

**Here tuning using Grid Search was applied on the 3 models, then the 3 kinds of data (raw, normalized, and smoothed) were processed**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV, KFold, cross_val_predict
from sklearn.cross_decomposition import PLSRegression
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, make_scorer
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')

# Prepare the datasets:
# X_raw: using the original spectral data from 'data'
# X_norm: using the normalized data from 'data_normalized'
# X_smooth: using the smoothed data from 'data_smoothed'
spectral_cols = data.columns[1:-1]  # assuming first col is Sample ID and last is Moisture

X_raw = data[spectral_cols].values
X_norm = data_normalized[spectral_cols].values
X_smooth = data_smoothed[spectral_cols].values
y = data['Moisture'].values

# Set up 5-fold cross-validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Define a RMSE scorer (GridSearchCV minimizes the score, so we use negative RMSE)
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
rmse_scorer = make_scorer(rmse, greater_is_better=False)

# Define parameter grids for each model:
pls_param_grid = {'n_components': [5, 10, 15, 20]}
svr_param_grid = {'svr__C': [1, 10, 100], 'svr__gamma': [0.001, 0.01, 0.1]}
ann_param_grid = {
    'mlp__hidden_layer_sizes': [(50,), (100,), (50,50), (100,50)],
    'mlp__alpha': [0.0001, 0.001, 0.01]
}

# Create a dictionary to hold each preprocessed dataset
preproc_data = {
    'Raw': X_raw,
    'Normalized': X_norm,
    'Smoothed': X_smooth
}

results = {}

for method, X_data in preproc_data.items():
    results[method] = {}
    print(f"--- Preprocessing: {method} ---")

    # PLS Regression Grid Search
    pls = PLSRegression()
    pls_grid = GridSearchCV(pls, pls_param_grid, cv=cv, scoring=rmse_scorer, refit=True)
    pls_grid.fit(X_data, y)
    best_pls_params = pls_grid.best_params_
    best_pls_rmse = -pls_grid.best_score_
    results[method]['PLS'] = {'params': best_pls_params, 'grid_RMSE': best_pls_rmse}
    print("PLS best params:", best_pls_params, "Grid RMSE:", best_pls_rmse)

    # Get full metrics using cross_val_predict for PLS
    y_pred_pls = cross_val_predict(pls_grid.best_estimator_, X_data, y, cv=cv)
    r2_pls = r2_score(y, y_pred_pls)
    rmse_pls = np.sqrt(mean_squared_error(y, y_pred_pls))
    results[method]['PLS'].update({'R2': r2_pls, 'RMSE': rmse_pls})
    print("PLS full metrics: R²:", r2_pls, "RMSE:", rmse_pls)

    # SVR Grid Search: using a pipeline with scaling and SVR
    svr_pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('svr', SVR(kernel='rbf'))
    ])
    svr_grid = GridSearchCV(svr_pipeline, svr_param_grid, cv=cv, scoring=rmse_scorer, refit=True)
    svr_grid.fit(X_data, y)
    best_svr_params = svr_grid.best_params_
    best_svr_rmse = -svr_grid.best_score_
    results[method]['SVR'] = {'params': best_svr_params, 'grid_RMSE': best_svr_rmse}
    print("SVR best params:", best_svr_params, "Grid RMSE:", best_svr_rmse)

    y_pred_svr = cross_val_predict(svr_grid.best_estimator_, X_data, y, cv=cv)
    r2_svr = r2_score(y, y_pred_svr)
    rmse_svr = np.sqrt(mean_squared_error(y, y_pred_svr))
    results[method]['SVR'].update({'R2': r2_svr, 'RMSE': rmse_svr})
    print("SVR full metrics: R²:", r2_svr, "RMSE:", rmse_svr)

    # ANN Grid Search: using a pipeline with scaling and MLPRegressor
    ann_pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('mlp', MLPRegressor(max_iter=1000, random_state=42))
    ])
    ann_grid = GridSearchCV(ann_pipeline, ann_param_grid, cv=cv, scoring=rmse_scorer, refit=True)
    ann_grid.fit(X_data, y)
    best_ann_params = ann_grid.best_params_
    best_ann_rmse = -ann_grid.best_score_
    results[method]['ANN'] = {'params': best_ann_params, 'grid_RMSE': best_ann_rmse}
    print("ANN best params:", best_ann_params, "Grid RMSE:", best_ann_rmse)

    y_pred_ann = cross_val_predict(ann_grid.best_estimator_, X_data, y, cv=cv)
    r2_ann = r2_score(y, y_pred_ann)
    rmse_ann = np.sqrt(mean_squared_error(y, y_pred_ann))
    results[method]['ANN'].update({'R2': r2_ann, 'RMSE': rmse_ann})
    print("ANN full metrics: R²:", r2_ann, "RMSE:", rmse_ann)

    print()

# Print the consolidated results:
print("Consolidated Results:")
for method, res in results.items():
    print(f"{method} Data:")
    for model, metrics in res.items():
        print(f"  {model}:")
        print(f"      Best Params: {metrics['params']}")
        print(f"      Grid RMSE: {metrics['grid_RMSE']:.3f}")
        print(f"      R²: {metrics['R2']:.3f}")
        print(f"      RMSE: {metrics['RMSE']:.3f}")
    print()

# --- Plotting Section for R² and RMSE Bar Charts ---
import matplotlib.pyplot as plt

methods_list = list(results.keys())  # e.g., ['Raw', 'Normalized', 'Smoothed']
model_list = ['PLS', 'SVR', 'ANN']

# Create arrays for R² and RMSE with dimensions: [n_methods x n_models]
r2_array = np.zeros((len(methods_list), len(model_list)))
rmse_array = np.zeros((len(methods_list), len(model_list)))

for i, method in enumerate(methods_list):
    for j, model in enumerate(model_list):
        r2_array[i, j] = results[method][model]['R2']
        rmse_array[i, j] = results[method][model]['RMSE']

# Plot R² bar chart (figure size (5,3))
fig, ax = plt.subplots(figsize=(5, 3))
bar_width = 0.25
x = np.arange(len(methods_list))
for j, model in enumerate(model_list):
    ax.bar(x + j*bar_width, r2_array[:, j], bar_width, label=model)
ax.set_xticks(x + bar_width)
ax.set_xticklabels(methods_list)
ax.set_ylabel("R² Score")
ax.set_title("R² Comparison Across Preprocessing Methods and Models")
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()

# Plot RMSE bar chart (figure size (5,3))
fig, ax = plt.subplots(figsize=(5, 3))
for j, model in enumerate(model_list):
    ax.bar(x + j*bar_width, rmse_array[:, j], bar_width, label=model)
ax.set_xticks(x + bar_width)
ax.set_xticklabels(methods_list)
ax.set_ylabel("RMSE")
ax.set_title("RMSE Comparison Across Preprocessing Methods and Models")
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()
```

# Applying Standard Normal Variate (SNV)  on raw data and then executing the same 3 ML models

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_decomposition import PLSRegression
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Set up 5-fold cross-validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Define spectral columns (assuming first col is 'Sample ID' and last is 'Moisture')
spectral_cols = data.columns[1:-1]

# Prepare feature arrays from the different preprocessing methods:
# Raw spectral data:
X_raw = data[spectral_cols].values

# Normalized data (assumed to be computed in data_normalized):
X_norm = data_normalized[spectral_cols].values

# Smoothed data (assumed to be computed in data_smoothed):
X_smooth = data_smoothed[spectral_cols].values

# Additional Preprocessing: Standard Normal Variate (SNV)
def snv_transform(row):
    return (row - row.mean()) / row.std()

data_snv = data.copy()
data_snv[spectral_cols] = data_snv[spectral_cols].apply(snv_transform, axis=1)
X_snv = data_snv[spectral_cols].values

# Target variable remains the same
y = data['Moisture'].values

# Create a dictionary for all preprocessing methods
preproc_data = {
    'Raw': X_raw,
    'Normalized': X_norm,
    'Smoothed': X_smooth,
    'SNV': X_snv
}

# Define models to compare
models = {
    'PLS': PLSRegression(n_components=10),  # You can adjust the number of components
    'SVR': make_pipeline(StandardScaler(), SVR(kernel='rbf', C=100, gamma=0.01)),
    'ANN': MLPRegressor(hidden_layer_sizes=(100,), activation='relu', solver='adam',
                        max_iter=1000, random_state=42)
}

# Dictionary to store performance results
results = {}

for method, X in preproc_data.items():
    results[method] = {}
    print(f"--- Preprocessing: {method} ---")
    for model_name, model in models.items():
        # Obtain cross-validated predictions
        y_pred = cross_val_predict(model, X, y, cv=cv)
        # Compute R² and RMSE
        r2 = r2_score(y, y_pred)
        rmse = np.sqrt(mean_squared_error(y, y_pred))
        results[method][model_name] = {'R2': r2, 'RMSE': rmse}
        print(f"{model_name}: R² = {r2:.3f}, RMSE = {rmse:.3f}")
    print()

# Prepare arrays for plotting
methods = list(results.keys())
model_names = list(models.keys())
r2_values = np.zeros((len(methods), len(model_names)))
rmse_values = np.zeros((len(methods), len(model_names)))

for i, method in enumerate(methods):
    for j, model_name in enumerate(model_names):
        r2_values[i, j] = results[method][model_name]['R2']
        rmse_values[i, j] = results[method][model_name]['RMSE']

# Plot R² comparison
fig, ax = plt.subplots(figsize=(5, 3))
width = 0.2  # width of each bar
x = np.arange(len(methods))
for j, model_name in enumerate(model_names):
    ax.bar(x + j*width, r2_values[:, j], width, label=model_name)
ax.set_xticks(x + width)
ax.set_xticklabels(methods)
ax.set_ylabel("R² Score")
ax.set_title("R² Comparison (Including SNV)")
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()

# Plot RMSE comparison
fig, ax = plt.subplots(figsize=(5, 3))
for j, model_name in enumerate(model_names):
    ax.bar(x + j*width, rmse_values[:, j], width, label=model_name)
ax.set_xticks(x + width)
ax.set_xticklabels(methods)
ax.set_ylabel("RMSE")
ax.set_title("RMSE Comparison (Including SNV)")
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()
```
